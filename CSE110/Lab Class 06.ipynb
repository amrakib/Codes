***Task 7***

```python
def nested_to_linear(dim_list , index, final_list):
    if len(dim_list) == index:
         return final_list
    elif type(dim_list[index]) != list:
         final_list.append(dim_list[index])
    else:
          nested_to_linear(dim_list[index], 0, final_list)
    return (nested_to_linear(dim_list, index + 1, final_list))

print(nested_to_linear(["start", 10, [4, 2, [11, [9, "mid", 3, [1, 0], 6]], 8], "Done"],   0 ,   []))
```
