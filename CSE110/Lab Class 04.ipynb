***Class Evaluation Task 2***

```python
# 2.0
var = input()
consonant = 0
vowel = 0
vowels = "AaEeIiOoUu"
for i in range(len(var)):
    if ("a" <= var[i] <= "z") or ("A" <= var[i] <= "Z"):
        if var[i] in vowels:
            vowel += 1
        else:
            consonant += 1
if (vowel > 0 and consonant > 0) and (vowel % 3 == 0 and consonant % 5 == 0):
    print("Aaarr! Me Plunder!!")
else:
    print("Blimey! No Plunder!!")
```

***Home Task 2***

```python
#task 2
var = input()
first = ""
for i in range(len(var)):
    if var[i] == ",":
        continue
    if var[i] == " ":
        break
    else:
        first += var[i]
second = var[i+1:]
last = ""
x = 0
y = 0
while x < len(first) or y < len(second):
    if x < len(first) and y < len(second):
        last += (first[x]+second[y])
        x += 1
        y += 1
    elif x < len(first) and y >= len(second):
        last += first[x]
        x += 1
    elif x >= len(first) and y < len(second):
        last += second[y]
        y += 1
print(last)
```

***Home Task 4***

```python
#task 4
var = input("Enter a string: ")
output = ""
upper = True
for i in var:
    if 'a' <= i <= 'z' or 'A' <= i <= 'Z':
        if upper:
            output += chr(ord(i) - 32) if 'a' <= i <= 'z' else i
        else:
            output += chr(ord(i) + 32) if 'A' <= i <= 'Z' else i
        upper = not upper
    else:
        output += i
print("Modified string:", output)
```
