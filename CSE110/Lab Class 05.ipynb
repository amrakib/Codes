```python
#CW-1.3
def is_pos(num):
    return num > 0

# sample calls are given below-
result = is_pos(-5)
print( result )
result = is_pos(12)
print( result )
```

```python
#CW-1.4
def sequence(n):
    if is_pos(n):
        for i in range(0, n+1, 1):
            if is_even(i):
                print(i, end=" ")
    else:
        for i in range(n, 0, 1):
            if not is_even(i):
                print(i, end=" ")

# sample calls are given below-
sequence(10)
sequence(-7)
sequence(7)
sequence(-8)
```

```python
#CW-2.1
def is_valid_triangle(side1, side2, side3):
    if (side1 + side2 > side3) and (side2 + side3 > side1) and (side3 + side1 > side2):
        return True
    else:
        return False

# sample calls are given below-
result = is_valid_triangle(7, 5, 10)
print(result)
result = is_valid_triangle(3, 2, 1)
print(result)
```

```python
#CW-2.2
import math
def tri_area(a, b, c):
    if is_valid_triangle(a, b, c):
        s = (a + b + c) / 2
        circle_area = math.sqrt(s * (s - a) * (s - b) * (s - c))
        print(round(circle_area,3))
    else:
        print("Canâ€™t form triangle")

# sample calls are given below-
tri_area(3,2,1)
tri_area(7,5,10)
```

***Class Task 3***

```python
#CW-3.1
def circle(radius):
    circle_area = math.pi * math.pow(radius, 2)
    return round(circle_area, 4)

# sample calls are given below-
circle(5)
```

```python
#CW-3.2
def sphere(radius):
    circle_volume = (4/3) * math.pi * math.pow(radius, 3)
    return round(circle_volume, 4)

# sample calls are given below-
sphere(5)
```

```python
#CW-3.3
def fitting(diameter1, diameter2, dimension):
    if diameter1 != diameter2:
        if dimension == 3:
            circle_volume1 = sphere(diameter1 / 2)
            circle_volume2 = sphere(diameter2 / 2)
            if circle_volume1 > circle_volume2:
                volume_left = circle_volume1 - circle_volume2
                print(f"Sphere-2 can fit inside Sphere-1 and {round(volume_left, 3)} cubic units would be left.")
            elif circle_volume2 > circle_volume1:
                volume_left = circle_volume2 - circle_volume1
                print(f"Circle-1 can fit inside Circle-2 and {round(volume_left, 3)} square units would be left.")
        elif dimension == 2:
            circle_area1 = circle(diameter1 / 2)
            circle_area2 = circle(diameter2 / 2)
            if circle_area1 > circle_area2:
                area_left = circle_area1 - circle_area2
                print(f"Circle-2 can fit inside Circle-1 and {round(area_left, 3)} square units would be left.")
            elif circle_area2 > circle_area1:
                area_left = circle_area2 - circle_area1
                print(f"Circle-1 can fit inside Circle-2 and {round(area_left, 3)} square units would be left.")
    else:
        print("Impossible to fit.")

# sample calls are given below-
fitting(8,10,2)
fitting(30,14,3)
fitting(5,5,3)
```

```python
#HW-1.1
def is_prime(num):
    div_count = 0
    for i in range(1, num+1):
        if num % i == 0:
            div_count += 1
    return div_count == 2

# sample calls are given below-
prime_check = is_prime(7)
print(prime_check)
prime_check = is_prime(15)
print(prime_check)
```

```python
#HW-1.2
def is_perfect(num):
    div_sum = 0
    for i in range(1, num):
        if num % i == 0:
            div_sum += i
    return div_sum == num

# sample calls are given below-
perfect_check = is_perfect(6)
print(perfect_check)
perfect_check = is_perfect(33)
print(perfect_check)
```

```python
#HW-1.3
num = int(input())
def special_sum(num):
    sum = 0
    for i in range(1, num+1):
        if is_prime(i) or is_perfect(i):
            sum += i
    return sum

# sample calls are given below-
result = special_sum(num)
print(result)
```

```python
#HW-2.1
def show_dots(num):
    for i in range(num):
        print(".", end="")

# sample calls are given below-
show_dots(5)
show_dots(3)
```

```python
#HW-2.2
def show_palindrome(num):
    for i in range(1, num+1):
        print(i, end="")
    for i in range(num-1, 0, -1):
        print(i, end="")

# sample calls are given below-
show_palindrome(5)
```

```python
#HW-2.3
def show_triangle(n):
    for i in range(1, n+1):
        show_dots(n-i)
        show_palindrome(i)
        show_dots(n-i)
        print()

# sample calls are given below-
show_triangle(5)
```

```python
#HW-3.1
def calc_tax(age, salary):
    if age < 18 or salary < 10000:
        return 0
    elif 10000 <= salary <= 20000:
        return int((salary*7/100))
    elif salary < 20000:
        return (salary*14/100)

# sample calls are given below-
t = calc_tax(20,18000)
print(t)
```

```python
#HW-3.2
def calc_yearly_tax():
    age = int(input())
    total_tax = 0
    for i in range(1,13):
        salary = int(input(f'Month number {i}:'))
        tax = int(calc_tax(age, salary))
        print(f'Month{i} tax: ',tax)
        total_tax += tax
    print(f'Total Yearly Tax: {total_tax}')

# sample calls are given below-
calc_yearly_tax()
```

```python
#HW-5.0
N = int(input())
def sequence_iterative(N):
    result = 0
    sign = 1
    for i in range(1, N+1):
        result += sign * i
        sign *= -1
    return result

# sample calls are given below-
sequence_iterative(N)
```

```python
#HW-5.1
N = int(input())
def sequence_recursive(N):
    if N == 1:
        return 1
    else:
        if N % 2 == 0:
            return sequence_recursive(N-1) - N
        else:
            return sequence_recursive(N-1) + N

# sample calls are given below-
sequence_recursive(N)
```
