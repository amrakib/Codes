***Class Evaluation Task 3***

```python
# Class Evaluation Task 3
distance = int(input())
seconds = int(input())
speed = (distance/1000)/(seconds/3600)
if speed < 60:
    print(f"{speed} km/h")
    print("Too slow. It needs more changes.")
elif 60 <= speed <= 90:
    print(f"{speed} km/h")
    print("Velocity is okay. The car is ready!")
elif speed > 90:
    print(f"{speed} km/h")
    print("Too fast. Only a few changes should suffice.")
```
