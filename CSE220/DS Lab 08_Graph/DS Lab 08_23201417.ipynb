```python
import numpy as np
class Node:
    def __init__(self, elem, next=None):
        self.elem = elem
        self.next = next
    
def printLinkedList(head):
    temp = head
    while temp != None:
        if temp.next != None:
            print(temp.elem, end='-->')
        else:
            print(temp.elem)
        temp = temp.next
    
# Creating a Graph
class Graph:
    def __init__(self, vertices):
        self.vertices = vertices
        self.adjacentMatrix = np.zeros((self.vertices, self.vertices), dtype=int)
        self.adjacentList = np.array([Node(i) for i in range(vertices)],dtype=object)
        

    def undirected_matrix_connections(self, start, end):
        self.adjacentMatrix[start][end] = 1
        self.adjacentMatrix[end][start] = 1

    def undirected_weighted_matrix_connections(self, start, end, weight):
        self.adjacentMatrix[start][end] = weight
        self.adjacentMatrix[end][start] = weight

    def directed_matrix_connections(self, start, end):
        self.adjacentMatrix[start][end] = 1

    def directed_weighted_matrix_connections(self, start, end, weight):
        self.adjacentMatrix[start][end] = weight

    def undirected_list_connections(self, start, end):
        head = self.adjacentList[start]
        temp = head.next
        head.next = Node(end)
        head.next.next = temp
        head = self.adjacentList[end]
        temp = head.next
        head.next = Node(start)
        head.next.next = temp

    def undirected_weighted_list_connections(self, start, end_and_weight):
        head = self.adjacentList[start]
        temp = head.next
        head.next = Node(end_and_weight)
        head.next.next = temp
        head = self.adjacentList[end_and_weight[0]]
        temp = head.next
        head.next = Node((start, end_and_weight[1]))
        head.next.next = temp

    def directed_list_connections(self, start, end):
        head = self.adjacentList[start]
        temp = head.next
        head.next = Node(end)
        head.next.next = temp

    def directed_weighted_list_connections(self, start, end_and_weight):
        head = self.adjacentList[start]
        temp = head.next
        head.next = Node(end_and_weight)
        head.next.next = temp
```

```python
#Task 1 (Adjacent Matrix)
def find_max_degree_adjacent_matrix(reference):
    max_degree = None
    max = 0
    for row in range(reference.vertices):
        sum = 0
        for col in range(reference.vertices):
            if reference.adjacentMatrix[row][col] != 0:
                sum += 1
        if sum > max:
            max_degree = Node(row)
            max = sum
        elif sum == max:
            another_max = Node(row, max_degree)
            max_degree = another_max
    return max, max_degree

max_vertex, vertex = find_max_degree_adjacent_matrix(undirected)
print("Maximum Degree:", max_vertex)
print("Vertex/Vertexes that have the maximum degree:", end =" ")
printLinkedList(vertex)
```

```python
#Task 4 (Adjacent Matrix)
def directed_weighted_matrix_to_undirected(reference):
    for i in range(reference.vertices):
        for j in range(reference.vertices):
            if reference.adjacentMatrix[i][j] != 0:
                reference.adjacentMatrix[j][i] = reference.adjacentMatrix[i][j]
    return reference

print("Directed Weighted Adjacent Matrix:\n", directed_weighted.adjacentMatrix, end="\n\n")
print("After Conversion:\n", directed_weighted_matrix_to_undirected(directed_weighted).adjacentMatrix)
```
